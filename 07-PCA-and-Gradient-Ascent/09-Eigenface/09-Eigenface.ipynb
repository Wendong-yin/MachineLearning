## 特征脸

![IMAGE](https://farm1.staticflickr.com/876/42122699804_ea756081fa_o.png)

X 的第一行和 W 的第一行进行点乘，得到的是原来数据在 W 方向上的映射大小。我们可以换一个角度来向：我们可以把 W 也想象成一个样本，它包含了k 个方向，**第一行是最重要的方向**，第二行是次重要的方向。

X 的每一行是一张脸，W 的每一行也是一张脸，叫做特征脸(agent face)。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_lfw_people
```

```python
faces = fetch_lfw_people()
```

```python
faces.keys()
```

```python
faces.data.shape 
```

```python
faces.target_names
```

```python
faces.images.shape # 每一张图片是62*47
```

```python
random_indexes = np.random.permutation(len(faces.data))
X = faces.data[random_indexes]
example_faces = X[:36,:] # 前36张脸
example_faces.shape
```

```python
def plot_faces(faces):
    
    fig, axes = plt.subplots(6, 6, figsize=(10, 10),
                         subplot_kw={'xticks':[], 'yticks':[]},
    gridspec_kw=dict(hspace=0.1, wspace=0.1)) 
    for i, ax in enumerate(axes.flat):
        ax.imshow(faces[i].reshape(62, 47), cmap='bone')
    plt.show()
    
plot_faces(example_faces)
```

### 特征脸

```python
%%time
from sklearn.decomposition import PCA 
# 用随机的方式求解 pca
pca = PCA(svd_solver='randomized')
pca.fit(X)
```

```python
pca.components_.shape
```

```python
plot_faces(pca.components_[:36,:])
```
