## 什么是多项式回归

![IMAGE](https://farm2.staticflickr.com/1755/42841143731_7f9e2a2a56_o.png)
有些数据用非线性拟合效果会更好。
如果我们把$x^{2}$ 看做是x 的另一个特征，那么其实其实本质还是线性回归。
> PCA 是的本质是降维，而多项式回归的本质是升维。这种思路，在之后的 SVM 算法中还会提现。

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, 100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 线性回归？

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict, color='r')
plt.show()
```

### 解决方案， 添加一个特征

```python
X2 = np.hstack([X, X**2])
```

```python
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
# 我们要将 x 排序之后，再进行绘制
plt.plot(x,y, color='r') 
plt.show()
```

```python
plt.scatter(x, y)
# 我们要将 x 排序之后，再进行绘制
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r') 
plt.show()
```
